```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, fft, fftfreq
from scipy.signal import spectrogram
```

```python
fsk = np.fromfile('fsk_clean_passband_raw.i16', dtype=np.int16)
qpsk = np.fromfile('qpsk_constellation_clean_raw.i16', dtype=np.int16)
```

```python
fs = 48000.0 # sampling Frequency
fsk = np.fromfile('fsk_clean_bfsk_passband_raw.i16', dtype=np.int16)
x = fsk / 32768.0  #  float   

nperseg = 512 # FFT window length in spectrogram
noverlap = 256 # overlap sample count in spectrogram
######################################################################
# * nperseg      
# * noverlap     
######################################################################


f, t, S = spectrogram(x, fs=fs, nperseg=nperseg, noverlap=noverlap)
# f =  
# t =  
# S =  

plt.figure(figsize=(8,5))
plt.pcolormesh(t, f, 20*np.log10(S+1e-12), shading='gouraud', cmap='gray')
plt.ylim(9000, 15000)  # BFSK   
plt.colorbar(label='dB')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [Hz]')
plt.title('BFSK Spectrogram (Zoomed)')
plt.show()
```
