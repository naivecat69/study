```python
data = np.fromfile('qpsk_complex_int8_iq_interleaved_fs48000_rs6000_rrc035_sps8_len2s.raw', dtype = np.int8)
```

```python
data
```

```python
iq = data.reshape(-1,2)
I = iq[:,0].astype(np.float32)
Q = iq[:,1].astype(np.float32)
z = I + 1j*Q
```

```python
sps = 8 # samples per symbol
z_sym = z[sps//2 :: sps]
```

```python
plt.figure(figsize=(6,6))
plt.scatter(z_sym.real, z_sym.imag, s=10, alpha=0.5)
plt.axhline(0, color='k'); plt.axvline(0, color='k')
plt.title("QPSK Constellation (symbol samples)")
plt.xlabel("I"); plt.ylabel("Q")
plt.axis("equal"); plt.grid(True)
plt.show()
```

```python
count = int(input())
fibo_list = list()

while len(fibo_list) < count:
    temp = int(input())
    fibo_list.append(temp)

def fibonacci(n, dic):
    if n in dic:
        if n == 0:
            dic['zero_count'] = dic['zero_count'] + 1
            return dic['zero_count']
        elif n == 1:
            dic['one_count'] = dic['one_count'] + 1
            return dic['one_count']
        return dic[n]
    if n < 2:
        dic[n] = n
    else:
        dic[n] = fibonacci(n-1, dic) + fibonacci(n-2, dic)

    return dic[n]

for i in fibo_list:
    dic = {'one_count': 0, 'zero_count': 0}
    fibonacci(i, dic)
    print(dic['zero_count'], dic['one_count'])

```
